```python
## Train test split 
## search on google scikit learn train test split 
from sklearn.model_selection import train_test_split
```

```python
## X_train and Y_train---> used to train the model
## X_test  and Y_test -----> used to test the model
X_train, X_test, y_train, y_test = train_test_split(
   x, y, test_size=0.30, random_state=42)
## random_state is optional
```

```python
## standard the independent data
## use sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_train
```

```python
X_test=scaler.transform(X_test)
X_test
```

```python
## to get orginal fro we use
## scaler.inverse_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
## we have multiple independent features so this is our multilinear regression

## cross validation
from sklearn.model_selection import cross_val_score

```

```python
## regression objection using linear regressio
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse = cross_val_score(regression, X_train, y_train, scoring="neg_mean_squared_error",cv=10)
## cv means ---> 5 diffrenent model with different mean squared error
```

```python
np.mean(mse)
```

```python
## prediction on test data
reg_predict=regression.predict(X_test)
reg_predict ## predicted value
```

```python
## check weather predict is correct or not
import seaborn as sns
sns.displot(reg_predict-y_test,kind="kde")
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_predict,y_test)
```
